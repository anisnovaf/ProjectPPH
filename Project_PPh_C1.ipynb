## Data analysis

### Subtask:
Analyze the cleaned and wrangled financial data.


**Reasoning**:
Analyze trends and patterns in key financial metrics, calculate and analyze key financial ratios, investigate correlations, and analyze the impact of tax holidays.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Analyze trends in key financial metrics
plt.figure(figsize=(15, 5))
plt.plot(df['tahun'], df['pendapatan_penjualan'], label='Revenue', marker='o', color='blue')
plt.plot(df['tahun'], df['total_hpp'], label='Cost of Goods Sold', marker='o', color='green')
plt.plot(df['tahun'], df['laba_bersih_normal'], label='Net Profit', marker='o', color='red')
plt.xlabel('Year')
plt.ylabel('Amount (IDR)')
plt.title('Trends in Key Financial Metrics')
plt.legend()
plt.grid(True)
plt.show()


# 2. Calculate and analyze key financial ratios
plt.figure(figsize=(15, 5))
plt.plot(df['tahun'], df['gross_profit_margin'], label='Gross Profit Margin', marker='o', color='purple')
plt.plot(df['tahun'], df['operating_profit_margin'], label='Operating Profit Margin', marker='o', color='orange')
plt.plot(df['tahun'], df['net_profit_margin'], label='Net Profit Margin', marker='o', color='brown')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.title('Trends in Profitability Ratios')
plt.legend()
plt.grid(True)
plt.show()


# 3. Investigate correlations between financial metrics
correlation_matrix = df[['pendapatan_penjualan', 'total_hpp', 'laba_bersih_normal', 'gross_profit_margin', 'operating_profit_margin', 'net_profit_margin']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Financial Metrics')
plt.show()

# Scatter Plots showing relationship between revenue and profit.
plt.figure(figsize=(8, 6))
plt.scatter(df['pendapatan_penjualan'], df['laba_bersih_normal'], color='skyblue')
plt.xlabel('Revenue')
plt.ylabel('Net Profit')
plt.title('Revenue vs. Net Profit')
plt.grid(True)
plt.show()


# 4. Analyze the impact of tax holidays
tax_holiday_impact = df.groupby('tax_holiday_status')['laba_bersih_normal'].mean()
print(tax_holiday_impact)
plt.figure(figsize=(8, 6))
tax_holiday_impact.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.ylabel('Average Net Profit')
plt.title('Impact of Tax Holidays on Average Net Profit')
plt.show()
```
